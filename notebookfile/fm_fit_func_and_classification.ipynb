```python
from lib import fm, build_data
```

# FM Fit sin function

```python
x_train_data, y_train_data = build_data.build_sin()

w_0, w, v = fm.train(x_train_data, y_train_data, 30, 50, 'regr')
y_results = fm.predict(w_0, w, v, x_train_data, 'regr')

fm.show_result('regr', x_train_data, y_train_data, y_results)
```

# FM Fit linear Function

```python
x_train_data, y_train_data = build_data.build_line()

w_0, w, v = fm.train(x_train_data, y_train_data, 30, 100, 'regr')

y_results = fm.predict(w_0, w, v, x_train_data, 'regr')
fm.show_result('regr', x_train_data, y_train_data, y_results)
```

# FM 2 classification

```python
label_train, features_train = build_data.build_iris("./data/iris_2class.txt")
w_0, w, v = fm.train(features_train, label_train, 30, 100, '2_class')

y_results = fm.predict(w_0, w, v, features_train, '2_class')
fm.show_result('2_class', x_train_data, label_train, y_results)
```
